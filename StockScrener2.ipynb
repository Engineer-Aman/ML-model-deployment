```python
import yfinance as yf
import talib
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Function to fetch historical stock data for a given time frame
def fetch_stock_data(stock_symbol, period='1d', interval='1d'):
    data = yf.download(stock_symbol, period=period, interval=interval)
    return data

# Function to calculate 44-period Moving Average and RSI
def calculate_indicators(data):
    data['SMA_44'] = talib.SMA(data['Close'], timeperiod=44)  # 44-period Simple Moving Average
    data['RSI'] = talib.RSI(data['Close'], timeperiod=14)  # 14-period RSI (can be customized)
    return data

# Function to scan stocks based on criteria
def scan_stocks(stocks, time_frame, moving_average_condition, rsi_upper, rsi_lower):
    filtered_stocks = []

    for stock in stocks:
        data = fetch_stock_data(stock, period='1mo', interval=time_frame)
        data = calculate_indicators(data)

        # Get the last value of the 44-period moving average and RSI
        last_sma_44 = data['SMA_44'].iloc[-1]
        last_rsi = data['RSI'].iloc[-1]
        previous_sma_44 = data['SMA_44'].iloc[-2]

        # Apply the screening conditions
        if moving_average_condition == "rising" and last_sma_44 > previous_sma_44:
            if last_rsi >= rsi_upper:
                filtered_stocks.append(stock)
        elif moving_average_condition == "falling" and last_sma_44 < previous_sma_44:
            if last_rsi <= rsi_lower:
                filtered_stocks.append(stock)

    return filtered_stocks

# Function to plot the stock chart
def plot_stock_chart(stock_symbol, time_frame='1d'):
    data = fetch_stock_data(stock_symbol, period='1mo', interval=time_frame)
    data = calculate_indicators(data)

    fig = go.Figure(data=[go.Candlestick(x=data.index,
                                        open=data['Open'],
                                        high=data['High'],
                                        low=data['Low'],
                                        close=data['Close'],
                                        name="Price"),
                         go.Scatter(x=data.index, y=data['SMA_44'], mode='lines', name='44-period SMA'),
                         go.Scatter(x=data.index, y=data['RSI'], mode='lines', name='RSI')])

    fig.update_layout(title=f'{stock_symbol} Stock Chart', xaxis_title='Date', yaxis_title='Price', template='plotly_dark')
    fig.show()

# User input for parameters
stocks = ['RELIANCE.NS', 'TCS.NS', 'INFY.NS']  # Example Nifty stocks, replace with full list
time_frame = '1d'  # 1-day candles (could be changed to '1m', '5m', '1wk', etc.)
moving_average_condition = 'rising'  # 'rising' or 'falling'
rsi_upper = 60
rsi_lower = 40

# Scan the stocks
filtered_stocks = scan_stocks(stocks, time_frame, moving_average_condition, rsi_upper, rsi_lower)

# Show the filtered stocks
print("Stocks that meet the conditions:")
for stock in filtered_stocks:
    print(stock)

# Plot chart for a specific stock
plot_stock_chart('RELIANCE.NS', time_frame)  # Example: Plot for Reliance
```
